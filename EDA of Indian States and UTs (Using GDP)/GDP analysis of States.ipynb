```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('dark_background')
#Reading in the data
gdp_data=pd.read_csv("ab40c054-5031-4376-b52e-9813e776f65e.csv")
gdp_data
```

```python
#Storing the % growth separately
gdp_growth=gdp_data[gdp_data['Items  Description']=='(% Growth over previous year)']
gdp_growth
```

```python
#Taking a look at the final df
gdp_states_growth
```

Now to find the average GDP groeth on a state-wise basis.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gdp_states_tp=gdp_states_growth[(gdp_states_growth['Duration']!='2012-13')] #Removing the unwanted row
state_avg=gdp_states_tp.mean().sort_values(ascending=False)#State wise GDP growth rates
state_avg
```

```python
#To plot the Average growth rates of GDP for the states
plt.figure(figsize=(10, 8))
sns.barplot(x=state_avg, y=state_avg.index)
plt.xlabel('Average Growth Rate (GDP)')
plt.savefig('Average Growth Rates (GDP).png',bbox_inches='tight') #Saving the plot as a jpg
plt.show()
```

### Conclusions 
- Looking at the above figure it’s apparent that states in the North-east region of India have been consistently growing fast. In fact, the top 5 states with the highest average growth rates over the aforementioned years are all form North-East India, these are – Mizoram, Tripura, Nagaland, Manipur and Arunachal Pradesh.
- It’s interesting to note that 4 (Mizoram, Tripura, Nagaland and Manipur) of the top 5 states belong to the so called 7 sister states of India.
- The states with the lowest average GDP growth rate include – Goa, Meghalaya, Odisha, Sikkim and Jammu & Kashmir.

My home-state of Maharashtra has an average GDP growth rate of 11.26% which is slightly below the average All-India GDP of 11.8675% (obtained by taking the average of ‘All_India GDP’ column) but it’s also way below the median GDP growth rate of the states i.e. 12.22%.

It’s important to note that the bigger states (the one’s which contribute more GDP than others) will have more of an influence on the All India GDP growth rate.

- Identify the top-5 and the bottom-5 states based on GDP per capita.
- Find the ratio of highest per capita GDP to the lowest per capita GDP.


```python
#Finding the file path that the jupyter notebook is in
import os 
mypath=os.path.dirname(os.path.realpath('GDP Assignment Code.ipynb'))+"/1 B"
```

```python
#Reading all the necessary data 
#First finding all the csv files in the dir
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
onlyfiles
```

```python
#Extracting all statenames
states=list(map(lambda x: x.split('-')[1],onlyfiles))
states
```

```python
#Reading all the files as df's
#Code to read individual files, save only the data of the year 2014-2015 and merge them based on their rows 
#Made a looping construct to achieve the above goal
GDSP_all_1415=pd.DataFrame()
GDSP_all_1415
for i in range(len(onlyfiles)):
    NAD1=pd.read_csv(mypath+"/"+onlyfiles[i],index_col='Item',encoding="unicode_escape")
    x=NAD1['2014-15'].to_frame()
    x.columns=[states[i]]
    if GDSP_all_1415.empty:
        GDSP_all_1415=x
    else:
        GDSP_all_1415=pd.merge(GDSP_all_1415, x, how='inner', on='Item')
```

```python
GDSP_all_1415
```

```python
#To identify the top 5 and bottom 5 states based on GDP per capita
GDP_percapita=GDSP_all_1415.loc['Per Capita GSDP (Rs.)',].sort_values(ascending=False)
print("Top 5 states with GDP per capita =")
print(GDP_percapita.head(5),"\n")
print("Bottom 5 states with GDP per capita =")
print(GDP_percapita.tail(5),"\n")
print("Ratio of the state with the highest per capita GDP to the lowest per capita GDP is =",GDP_percapita['Goa']/GDP_percapita['Bihar'])
print("This means that Goa has a GDP per capita i.e approximately 8 times that of Bihar")
```

```python
#Plotting the GDP per capita for all the states-
GDP_percapita
plt.figure(figsize=(10, 8))
sns.barplot(x=GDP_percapita, y=GDP_percapita.index)
plt.xlabel('GDP_percapita in Rs. (2014-2105)')
plt.savefig('GDP per capita (2014-2015).png',bbox_inches='tight') #Saving the plot as a jpg
plt.show()
```

Plot the percentage contribution of primary, secondary and tertiary sectors as a percentage of total GDP for all the states.

```python
#Saving the percentage contribution of sectors for each state as a separate df
p=(GDSP_all_1415.loc[['Primary','Secondary','Tertiary']]/GDSP_all_1415.loc['Gross State Domestic Product'])*100
```

```python
#To broadly vizualize the contribution of Primary, Secondary and Tertiary sectors in all the states.
#Plotted a hierarchically clustered heatmap (along the rows)
#It has broadly formed three clusters, the states which have more contribution from secondary and tertiary sectors, 
#The second cluster has a lot of subdivisions but mostly contains states with high tertiary and primary sectors
#The third cluster has states whose GDSP is most contributed by the tertiary sector.
sns.clustermap(p.transpose(),cmap="Blues",col_cluster=False,figsize=(12,14))
plt.xlabel('Percentage contribution of the sectors to the total GDP')
plt.savefig('Percentage contribution of the sectors to the total GDP.png',bbox_inches='tight') #Saving the plot as a jpg
plt.show()
```

Used a hierarchically clustered heatmap (along the columns) to see if there are any emerging trends in the data.
- It has broadly formed three clusters; the first cluster has states which have more contribution from secondary and tertiary sectors (the first 5 states of the above plot).
- The second cluster has a lot of subdivisions but mostly contains states with high tertiary and primary sectors.
- The third cluster has states whose GDSP is most contributed by the tertiary sector (the last 10 states).

## Categorizing the data into C1, C2, C3, C4 based on their GDP per capita.

```python
state_categories=pd.qcut(GDSP_all_1415.loc['Per Capita GSDP (Rs.)'], q=[0,0.20,0.5, 0.85, 1],labels=["C4", "C3", "C2","C1"])
state_categories=state_categories.to_frame().transpose()
state_categories.set_index(np.array(['Categories']),inplace=True) # Getting a keyerror when not coerced to an np array
GDSP_all_1415_c=GDSP_all_1415.append(state_categories)
GDSP_all_1415_c
```

For each category C1, C2, C3, C4:
Find the top 3/4/5 sub-sectors (such as agriculture, forestry and fishing, crops, manufacturing etc.) which contribute to approx. 80% of the GSDP of each category






```python
#Cleaning up the data a bit and storing it in a new variable
GDSP_all_1415_sub=GDSP_all_1415_c.drop(['Primary','Secondary','Tertiary','TOTAL GSVA at basic prices',"Population ('00)",'Per Capita GSDP (Rs.)'])
#Selecting all sub sectors (while ignoring the sub sub sectors)
GDSP_all_1415_sub=GDSP_all_1415_sub.loc[['Agriculture, forestry and fishing','Mining and quarrying','Manufacturing',
                                'Electricity, gas, water supply & other utility services','Construction','Trade, repair, hotels and restaurants',
                               'Transport, storage, communication & services related to broadcasting','Financial services','Real estate, ownership of dwelling & professional services',
                               'Public administration','Other services','Taxes on Products','Gross State Domestic Product','Categories']]
GDSP_all_1415_sub
```

```python
#Grouping all the sub-sectors by using the user created Categories (C1,C2,C3,C4)
category_agg=GDSP_all_1415_sub.transpose().groupby('Categories')[GDSP_all_1415_sub.index[:len(GDSP_all_1415_sub.index)-1]].sum().transpose()
category_agg
```

```python
GDSP_all_1415_sub.loc['Categories'].sort_values() #looking at which states belong to which categories
```

```python
#Finding the % contribution of each subcategory
percentage_sub=(category_agg/category_agg.loc['Gross State Domestic Product'])*100
```

```python
#A crude way to find how many sub sectors are contributing to ~80% of the GDP for each category
print(percentage_sub.C1.sort_values(ascending=False).head(7)[1:].sum())
print(percentage_sub.C2.sort_values(ascending=False).head(8)[1:].sum())
print(percentage_sub.C3.sort_values(ascending=False).head(8)[1:].sum())
print(percentage_sub.C4.sort_values(ascending=False).head(8)[1:].sum())

```

```python
#Finding the sub-sectors contributing to ~80% of the GDP for each Category.
#For C1
c1=percentage_sub.C1.sort_values(ascending=False)[1:]
print("The following sub-sectors contribute to ~80% of the GDP for C1 -")
print(list(c1.head(6).index),"\n")
#For C2
c2=percentage_sub.C2.sort_values(ascending=False)[1:]
print("The following sub-sectors contribute to ~80% of the GDP for C2 - ")
print(list(c2.head(7).index),"\n")
#For C3
c3=percentage_sub.C3.sort_values(ascending=False)[1:]
print("The following sub-sectors contribute to ~80% of the GDP for C3 - ")
print(list(c3.head(7).index),"\n")
#For C4
c4=percentage_sub.C4.sort_values(ascending=False)[1:]
print("The following sub-sectors contribute to ~80% of the GDP for C4 - ")
print(list(c4.head(7).index),"\n")
```

```python
#Now to plot the contribution of the subsectors in each category
#To me a bar chart conveys much more info visually than a pie chart hence I've plotted the percentages as bar charts
# adjust figure size
plt.figure(figsize=(10, 28))

# subplot 1: C1
plt.subplot(4, 1, 1)
sns.barplot(y=c1.index, x=c1)
plt.title("Percentage contribution of sub-sectors in C1")
plt.xlabel('Percentage of GDP')

# subplot 2: C2
plt.subplot(4, 1, 2)
sns.barplot(y=c2.index, x=c2)
plt.title("Percentage contribution of sub-sectors in C2")
plt.xlabel('Percentage of GDP')

# subplot 3: C3
plt.subplot(4, 1, 3)
sns.barplot(y=c3.index,x=c3)
plt.title("Percentage contribution of sub-sectors in C3")
plt.xlabel('Percentage of GDP')

# subplot 4: C4
plt.subplot(4, 1, 4)
sns.barplot(y=c4.index,x=c4)
plt.title("Percentage contribution of sub-sectors in C4")
plt.xlabel('Percentage of GDP')

plt.savefig('Percentage contribution of the sub-sectors.png',bbox_inches='tight') #Saving the plot as a jpg
plt.show()
```

## Part 2- GDP and Education Drop-out Rates
To investigate whether there is any relationship between per capita GDP with drop-out rates in education.

```python
#Reading in the data
dropout_rates=pd.read_csv("rs_session243_au570_1.1.csv",index_col="Level of Education - State")
dropout_rates.drop('Sl. No.',axis=1,inplace=True)
```

```python
#Taking a look at the data
print(dropout_rates.columns) #Looking at the column names
```

```python
#Cleaning the data (to include only the years 2014-15)
dropout_rates_1415= dropout_rates[['Primary - 2014-2015','Upper Primary - 2014-2015','Secondary - 2014-2015'
                                   ,'Senior Secondary - 2014-2015']]
dropout_rates_1415.transpose()
```

```python
#Create a new df with the GDSP (2014-2015) the the drop out rates of 2014-15
dropout_GDP=dropout_rates_1415.transpose().append(GDSP_all_1415.loc['Per Capita GSDP (Rs.)'])
```

```python
#Changing the state names to merge the 2 files
list(map(lambda x: x.split('-')[1],onlyfiles))
state_names=list(map(lambda x: x.replace('_',' '),list(GDSP_all_1415.loc['Per Capita GSDP (Rs.)'].index)))
state_names
GDP_percapita=GDSP_all_1415.loc['Per Capita GSDP (Rs.)']
GDP_percapita.index=state_names #Changing the index to the new state names
GDP_percapita=GDP_percapita.to_frame()
GDP_percapita
```

```python
dropout_rates_1415
```

```python
#Changing some column names to enable merging (spellings in the 2 dfs do not match)
new_columns=list(dropout_rates_1415.index)
new_columns[6]='Chhattisgarh'
new_columns[-3]='Uttarakhand'
new_columns
dropout_rates_1415.index=pd.Series(new_columns)#Convert the list to a series
dropout_rates_1415
```

```python
#Merging the GDP percapita with the dropout df
dropout_GDP=pd.merge(dropout_rates_1415,GDP_percapita,left_on=dropout_rates_1415.index,right_on=GDP_percapita.index)
```

```python
dropout_GDP.index=dropout_GDP.key_0
dropout_GDP.drop(['key_0','Senior Secondary - 2014-2015'],axis=1,inplace=True)
dropout_GDP
```

Now to see if dropout rates from schools have any correlation with the GDP per capita of the states.


```python
from statistics import median 
#Lets try using a scatterplot to see if scatterplot can determine the correlation
#Plotting each education category dropout rates as different groups.
ax =dropout_GDP.plot.scatter(s=50,x='Per Capita GSDP (Rs.)', y='Primary - 2014-2015', label='Primary',color='DarkBlue')
b=dropout_GDP.plot.scatter(s=50,x='Per Capita GSDP (Rs.)', y='Upper Primary - 2014-2015', color='DarkGreen', label='Upper Primary', ax=ax)
dropout_GDP.plot.scatter(s=50,x='Per Capita GSDP (Rs.)', y='Secondary - 2014-2015', color='Red', label='Secondary', ax=b,figsize=(10,10))
plt.ylabel('Dropout Rates for 2014-2015')
plt.axvline(median(dropout_GDP['Per Capita GSDP (Rs.)']))
plt.text(median(dropout_GDP['Per Capita GSDP (Rs.)'])+2000,20,'Median GDP per capita') # Drawing a vline which denotes the median GDP per capita
plt.savefig('Dropout Rates Vs Per Capita GDP.png',bbox_inches='tight') #Saving the plot as a jpg
plt.show()
```

It can be seen that in most of the states the Primary and Upper Primary dropout rates are quite low but they do not seem to affect the GDP per capita. In fact, most of the states with low primary and secondary dropout rates have a low GDP (below median) and a considerable number of states with very low Primary and Upper Primary dropout rates have a GDP per capita above the median.

Now looking at the Secondary dropout rates, it looks like they have a negative correlation with the GDP per capita.
Most of the states with high Secondary dropout rates have their GDP per capita below the median. Most of the states with Secondary dropout rates that are relatively low (below 20%) have their GDP per capita above the median.
Obviously outliers exist in both these cases but nevertheless a general trend is observed w.r.t the Secondary dropout rates.
